Extramos los datos de Google Trends con la biblioteca [pytrends](https://github.com/GeneralMills/pytrends) ([código fuente](https://github.com/GeneralMills/pytrends/blob/master/pytrends/request.py)) y representemos los datos con matplotlib.

```python
# =========== IMPORTS ===========
from pytrends.request import TrendReq # pip3 install pytrends
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib
matplotlib.rcParams['figure.figsize'] = (20.0, 10.0)  # Cambiar los tamaños de las figuras

print(pd.__version__)
print(matplotlib.__version__)
# ================================
```

```python
# ============= MANEJADOR DE PETICIONES ================

keywords = ["learn python", "learn javascript", "learn ruby", "learn c++", "learn c"]  # Palabras clave, máximo 5
#keywords = ["tuina", "reflexology", "reiki", "shiatsu", "osteopathy"]
#keywords = ["Dalai Lama", "Eckhart Tolle", "Pope Francisco", "Thich Nhat Hanh", "Rhonda Byrne"]

pytrend = TrendReq()
pytrend.build_payload(kw_list=keywords) # Puedes seleccionar geo="ES" para buscar en España

# ======================================================

# --------- Interés a lo largo del tiempo -----------
over_time = pytrend.interest_over_time() # Obtenemos los datos 

over_time.plot() # Dibujar el gráfico

# Activar grilla
plt.grid(True)

plt.rc("font", size=14) # Cambiar tamaño de los títulos y leyendas
plt.suptitle('Interés de las palabras clave "%s" en GoogleTrends' % ", ".join(keywords)) # Supratítulo

plt.rc("font", size=18) 
plt.title("Interés a lo largo del tiempo") # Título

plt.legend(loc=2) # Leyenda en la esquina superior izquierda

# Establecer los nombres de los ejes
plt.xlabel("Fechas") 
plt.ylabel("Interés")
# ----------------------------------------------------


# -------------  Interés por región ------------------
by_region = pytrend.interest_by_region()

matplotlib.rcParams['figure.figsize'] = (20.0, 60.0)
by_region.plot(kind="barh") # Gráfico de barras horizontal

 # Dar la vuelta a los índices para que aparezcan en orden alfabético 
plt.gca().invert_yaxis()  # <--- gca()  Get current axis
plt.ylabel("")
# ---------------------------------------------------


plt.show()
```
