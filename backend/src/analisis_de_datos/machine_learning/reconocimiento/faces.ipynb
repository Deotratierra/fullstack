## <center>Reconocimiento de rostros con OpenCV</center>

Para realizar el proceso de reconocimiento de rostros con OpenCV es necesario descargar el modelo preentrenado que puede encontrarse en [el reposiorio de OpenCV de Github](https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_alt.xml).

```python
# pip install opencv-python matplotlib

import cv2
from IPython.display import (Image, display)

# Instanciamos el detector de caras preentrenado
face_cascade = cv2.CascadeClassifier('models/haarcascade_frontalface_alt.xml')

def face_detector(img_path):
    """Detecta el número de rostros incluidos en la imagen
    cuya ruta es pasada como parámetro.
    
    :param img_path: Ruta a la imagen.
    :type img_path: str
    """
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray)
    return len(faces)

img_path = "img/10_faces.jpeg"
display(Image(img_path))

n_faces = face_detector(img_path)
print("Número de rostros detectados en la imagen: %d" % n_faces)
```
