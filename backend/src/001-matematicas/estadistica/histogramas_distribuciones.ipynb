# <center>Histogramas y distribuciones</center>

La representación más común de una distribución es un histograma, que es un gráfico que muestra la frecuencia o probabilidad de cada valor. El histograma muestra las frecuencias como un gráfico de barras que indica cuan frecuente un determinado valor ocurre en el conjunto de datos.

Las distribuciones se pueden clasificar en dos grandes grupos:

1. Las distribuciones continuas, que son aquellas que presentan un número infinito de posibles soluciones. Dentro de este grupo vamos a encontrar a las distribuciones:
    - [Normal](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_normal)
    - [Gamma](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_gamma)
    - [Chi cuadrado](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_%CF%87%C2%B2)
    - [t de Student](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_t_de_Student)
    - [Pareto](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_de_Pareto)
    
2. Las distribuciones discretas, que son aquellas en las que la variable puede tomar un número determinado de valores. Los principales exponentes de este grupo son las distribuciones:
    - [Poisson](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_de_Poisson)
    - [Binomial](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_binomial)
    - [Hipergeométrica](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_hipergeom%C3%A9trica)
    - [Bernoulli](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_de_Bernoulli)
    
[Referencia de distribuciones en scipy.stats](https://docs.scipy.org/doc/scipy/reference/stats.html)

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

Todas las distribuciones continuas en `scipy` están definidas en función de dos parámetros: `loc` y `scale`, que juegan distinto papel en función de la distribución que manejemos.

_____________________________________

### Distribución normal
Es una de las principales distribuciones, ya que es la que con más frecuencia aparece aproximada en los fenómenos reales. Tiene una forma acampanada y es simétrica respecto de un determinado parámetro estadístico.

```python
x = np.arange(-10, 10, 0.01) # Rango de -10 a 10 con saltos de 0.01

# Distribuciones normales
y1 = st.norm.pdf(x, -5, .5)  # Media = -5, desviación estándar = 0.5
y2 = st.norm.pdf(x, 0, 1)    # Media = 0, desviación estándar = 1
y3 = st.norm.pdf(x, 5, 2)    # Media = 2, desviación estándar = 2

# ------------------------------------------------------------------

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, y1, label="$\mu$ = -5 | $\sigma$ = 0.5")
ax.plot(x, y2, label="$\mu$ = 0  | $\sigma$ = 1")
ax.plot(x, y3, label="$\mu$ = 5  | $\sigma$ = 2")

plt.legend(loc=2, prop={'size': 12})
plt.show()
```

[Referencia de la función `norm`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm)

En esta distribución `loc` es la media y por tanto el centro de la distribución y `scale` es la desviación típica. Esta funciona aquí como una escala de concentración de los parámetros en torno a la media.
