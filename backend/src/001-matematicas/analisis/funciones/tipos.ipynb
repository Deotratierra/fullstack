### Funciones lineales
Representan líneas rectas oblicuas. Binomios de primer grado: <center> $f(x) = ax + b$</center>

- `a`: se le llama **pendiente** ya que determina la inclinación que tomará la línea del gráfico. 
- `b`: se le llama **ordenada al origen**, es el término independiente de la función.



```python
def funcion_lineal(x, a, b):
    return a*x + b

lineal = lambda x, a, b: a*x+b
```

>Cuando $a=0$, entonces $f(x) = b, por lo tanto se comporta como una [función constante](https://es.wikipedia.org/wiki/Funci%C3%B3n_constante).

#### Como sacar la función de una recta
Buscamos dos puntos en la recta que representen coordenadas enteras $(x, y) (x_1, y_1)$, en este caso $(-3, 6) (6, 0)$.


- **Forma punto pendiente** $\rightarrow y - y_1 = m(x - x_1)$
- **Forma pendiente ordenada al origen** $\rightarrow y = mx + b$
- **Forma estándar** $\rightarrow Ax + By = C$

![Tres formas de sacar la función de una gráfica lineal](img/3_formas_funcion_lineal.png)

```python
a, b = ((-3, 6), (6, 0))
coords = (a, b)
```

Primero hay que sacar la pendiente entre los dos puntos: $m = \frac{\Delta x}{\Delta y} = \frac{y_1-y}{x_1-x}$

```python
# Sólo números decimales
pendiente = lambda a, b: (b[1] - a[1]) / (b[0] - a[0])

# Con fracciones
def pendiente(a, b):
    from fractions import Fraction
    x, x_1, y, y_1 = (a[0], b[0], a[1], b[1])
    response = (y_1-y) / (x_1-x)
    if isinstance(response, float):
        return Fraction((y_1-y), (x_1-x))
    return response

print(pendiente((-7, -3) ,(-2, 4)))

pendiente(*coords)  # m
```

Luego aplicamos la fórmula con la cual queremos representarla.

```python
from sympy.abc import x, y

# Igualada a 0
def funcion_recta(a, b):
    f = lambda a, b: pendiente(a, b)*(x - a[0]) - y + a[1]
    return f(a, b)

pp = funcion_recta(*coords)
print(pp)                    # -2*x/3 - y + 4
print(str(pp)+" = 0")        # -2*x/3 - y + 4 = 0
```

### Funciones cuadráticas
Representan parábolas. Trinomios de segundo grado: <center>$f(x) = ax^2 + bx + c$</center>

- `c`: **ordenada al origen**.

```python
def funcion_cuadratica(x, a, b, c):
    return a*x**2 + b*x + c
```

Cuando $a>0$, la parábola es cóncava, cuando $a<0$ es convexa y cuando $a=0$, entonces $f(x) = c, por lo tanto se comporta como una [función constante](https://es.wikipedia.org/wiki/Funci%C3%B3n_constante).

#### Funciones cuadráticas en forma canónica o de vértice
Las funciones cuadráticas en forma de vértice tienen la forma $y = a(x-h)^2 + k$ donde el vértice de la parábola es e punto $(h, k)$.



____________________________________

### Funciones cúbicas
Representan curvas cúbicas. Cuatrinomio de tercer grado: <center>$f(x) = ax^3 + bx^2 + cx + d$</center>.

- `d`: **ordenada al origen**.

```python
def funcion_cubica(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d
```

___________________

## Representación 

```python
# Rango (eje X)
x = np.arange(-200, 201)

# -----------------------------

# Funciones:

#    Constante
y0 = funcion_lineal(x, 0, 300)

#    Lineal:
y1 = funcion_lineal(x, 1, 30) # Pendiente menor, ordenada al origen mayor
y2 = funcion_lineal(x, 2, 10) # Pendiente mayor, ordenada al origen menor
y3 = funcion_lineal(x, -3, 50) # Pendiente negativa, ordenada al origen mayor

#    Cuadrática
y4 = funcion_cuadratica(x, 0.03, 0, 0)

#    Cúbica
y5 = funcion_cubica(x, 0.0002, 0, 0, 0)

# -----------------------------

# Instanciamos figura
fig, ax = plt.subplots(figsize=(16, 8))
plt.ylim(-800, 800)  # Rango de y

# Dibujamos las fórmulas
ax.plot(x, y0, label="función constante")
ax.plot(x, y1, label="función lineal 1")
ax.plot(x, y2, label="función lineal 2")
ax.plot(x, y3, label="función lineal 3 (pendiente negativa)")
ax.plot(x, y4, label="función cuadrática")
ax.plot(x, y5, label="función cúbica")

plt.legend() # Leyenda
plt.grid()   # Grilla
plt.show()
```
