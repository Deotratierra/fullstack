# <center>Representación de funciones</center>

```python
import numpy as np
import matplotlib.pyplot as plt
```

____________________________________

### Funciones lineales
Representan líneas rectas oblicuas. Binomios de primer grado: <center> $f(x) = ax + b$</center>

- `a`: se le llama **pendiente** ya que determina la inclinación que tomará la línea del gráfico. 
- `b`: se le llama **ordenada al origen**, es el término independiente de la función.



```python
def funcion_lineal(x, a, b):
    return a*x + b
```

>Cuando $a=0$, entonces $f(x) = b, por lo tanto se comporta como una [función constante](https://es.wikipedia.org/wiki/Funci%C3%B3n_constante).

____________________________________

### Funciones cuadráticas
Representan parábolas. Trinomios de segundo grado: <center>$f(x) = ax^2 + bx + c$</center>

- `c`: **ordenada al origen**.

```python
def funcion_cuadratica(x, a, b, c):
    return a*x**2 + b*x + c
```

Cuando $a>0$, la parábola es cóncava, cuando $a<0$ es convexa y cuando $a=0$, entonces $f(x) = c, por lo tanto se comporta como una [función constante](https://es.wikipedia.org/wiki/Funci%C3%B3n_constante).

____________________________________

### Funciones cúbicas
Representan curvas cúbicas. Cuatrinomio de tercer grado: <center>$f(x) = ax^3 + bx^2 + cx + d$</center>.

- `d`: **ordenada al origen**.

```python
def funcion_cubica(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d
```

___________________

### Representación 

```python
# Rango (eje X)
x = np.arange(-100, 101) 

# -----------------------------

# Funciones:

#    Constante
y0 = funcion_lineal(x, 0, 300)

#    Lineal:
y1 = funcion_lineal(x, 1, 30) # Pendiente menor, ordenada al origen mayor
y2 = funcion_lineal(x, 2, 10) # Pendiente mayor, ordenada al origen menor
y3 = funcion_lineal(x, -3, 50) # Pendiente negativa, ordenada al origen mayor

#    Cuadrática
y4 = funcion_cuadratica(x, 0.03, 0, 0)

#    Cúbica
y5 = funcion_cubica(x, 0.0002, 0, 0, 0)
# -----------------------------

# Instanciamos figura
fig, ax = plt.subplots(figsize=(16, 8))

# Dibujamos las fórmulas
ax.plot(x, y0, label="funcion constante")
ax.plot(x, y1, label="funcion lineal 1")
ax.plot(x, y2, label="funcion lineal 2")
ax.plot(x, y3, label="funcion lineal 3 (pendiente negativa)")
ax.plot(x, y4, label="funcion cuadrática")
ax.plot(x, y5, label="funcion cúbica")

plt.legend() # Leyenda
plt.grid()   # Grilla
plt.show()
```
